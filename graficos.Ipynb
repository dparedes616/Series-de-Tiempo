## Graficos en Python

### Librerias

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Variables

```python
x = np.linspace(0, 100, 101)
y =np.sqrt(x)+5
```

### Graficos PLOT

```python
plt.plot(x, y, color="red")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Titulo del grafico")
plt.show()
```

### Graficos subplots

```python
plt.subplot(1,2,1)
plt.plot(x, y, color="red")
plt.title("Raiz de X")
plt.subplot(1,2,2)
plt.plot(y,x, color="blue")
plt.title("X al cuadrado")
plt.show()
```

### Graficos Orientados a Objetos

```python
grafico = plt.figure()
axe = grafico.add_axes([0,0,1,1])
axe.plot(x,y)
axe.set_xlabel("eje X")
axe.set_ylabel("eje Y")
axe.set_title("Titulo")
plt.show()
```

```python
grafico_2 = plt.figure()
axe = grafico_2.add_axes([0,0,1,1])
axe_2 = grafico_2.add_axes([0.4, 0.2, 0.5, 0.3])
axe.plot(x,y)
axe_2.plot(y,x)
axe.set_xlabel("eje X")
axe.set_ylabel("eje Y")
axe.set_title("Titulo")
axe_2.set_xlabel("eje X 2")
axe_2.set_ylabel("eje Y 2")
axe_2.set_title("Titulo 2")
plt.show()
```

### Subplots graficos orientados a objetos

```python
graficos,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
graficos,axes = plt.subplots(nrows=1, ncols=2)
for this_axes in axes:
    this_axes.plot(x, y)
```

```python
graficos,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("Grafico 1")
axes[1].set_title("Grafico 2")
plt.show()
```

```python
grafico_3 = plt.figure(figsize=(3,4), dpi=100)
axe = grafico_3.add_axes([0,0,1,1])
axe.plot(x,y)
plt.show()
```

```python
graficos,axes = plt.subplots(nrows=1, ncols=2,dpi=200, figsize=(9,3))
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("Grafico 1")
axes[1].set_title("Grafico 2")
plt.tight_layout()
plt.show()
```

```python
grafico_4 = plt.figure()
axes = grafico_4.add_axes([0,0,1,1])
axes.plot(x, x*2, label="2X")
axes.plot(x, x+2, label ="X+2")
axes.legend(loc=4)
plt.show()
```

```python
grafico_4.savefig("mi grafico.png", dpi=600)
grafico_4.savefig("mi grafico.jpg", dpi=600)
```

```python
grafico_4 = plt.figure()
axes = grafico_4.add_axes([0,0,1,1])
axes.plot(x, x*2, label="2X", color="red", lw=5, linestyle="dashed")
axes.plot(x, x+2, label ="X+2", color="#FF00FF", lw=5, alpha=0.3)
axes.legend(loc=0)
plt.show()
```

```python
x=np.linspace(0, 100, 21)
y=x**2
grafico_5 = plt.figure(figsize=(3,4), dpi=100)
axe = grafico_5.add_axes([0,0,1,1])
axe.plot(x,y, color="red", lw=0.5, marker= ".", markersize=10)
plt.show()
```
