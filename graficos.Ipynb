## Graficos en Python

### Librerias

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Variables

```python
x = np.linspace(0, 100, 101)
y =np.sqrt(x)+5
```

### Graficos PLOT

```python
plt.plot(x, y, color="red")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Titulo del grafico")
plt.show()
```

### Graficos subplots

```python
plt.subplot(1,2,1)
plt.plot(x, y, color="red")
plt.title("Raiz de X")
plt.subplot(1,2,2)
plt.plot(y,x, color="blue")
plt.title("X al cuadrado")
plt.show()
```
