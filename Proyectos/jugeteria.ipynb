### Librerias

```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

### Data

### Analisis ETE

```python
grafico_ventas = plt.figure()
ax = grafico_ventas.add_axes([0,0,1,1])
ax.plot(df_jugeteria["ventas"], label="Ventas")
ax.set_title("Inversi√≥n real en EEUU entre 2019-01 a 2023-12")
ax.set_xlabel("A√±o-Mes")
ax.set_ylabel("Miles de millones USD")
ax.legend(loc='best') 
ax.grid(True)
plt.show()
```

```python
grafico_ventas.savefig("Grafico de ventas.png", dpi=600 , bbox_inches="tight")
```

#### Analisis Adtitivo

```python
ete_aditive = seasonal_decompose(df_jugeteria["ventas"], model="aditive")
ete_aditive
```

```python
grafico_aditivo = ete_aditive.plot()
```

#### Analisis Multiplicativo

```python
ete_multiplicative = seasonal_decompose(df_jugeteria["ventas"], model="multiplicative")
ete_multiplicative
```

```python
grafico_multiplicativo = ete_multiplicative.plot()
```

```python
grafico_multiplicativo.savefig("Grafico de descomposici√≥n.png", dpi=600 , bbox_inches="tight")
```
