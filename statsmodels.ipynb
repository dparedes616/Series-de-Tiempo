```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
indice = pd.Index(sm.tsa.datetools.dates_from_range("1959Q1", "2009Q3"))
```

```python
datosmacro.index = indice
```

```python
datosmacro.head()
```

## Serie temporal de gastos por trimestre en los Estado Unidos

```python
data = datosmacro
data.head()
```

```python
grafico_1 = plt.figure(figsize=(10,5), dpi=300)
axe = grafico_1.add_axes([0,0,1,1])
axe= plt.plot(data["realcons"])
axe = plt.title("Gasto de consumo real en EEUU entre 1959-2009")
axe = plt.xlabel("AÃ±o-Trimestre")
axe = plt.ylabel("Gasto real en miles de millones USD")
axe =plt.grid(True)
plt.show()
```

## Serie temporal de inversiÃ³n real por trimestre en los Estado Unidos

```python
grafico_2 = plt.figure(figsize=(10,5), dpi=300)
axe = grafico_2.add_axes([0,0,1,1])
axe= plt.plot(data["realinv"])
axe = plt.title("Gasto de consumo real en EEUU entre 1959-2009")
axe = plt.xlabel("AÃ±o-Trimestre")
axe = plt.ylabel("Gasto real en miles de millones USD")
axe =plt.grid(True)
plt.show()
```

### Filtro Hodrick-Prescoot

Metodo para extraer la tendencia de la serie de tiempo

### Tencia de la inversiÃ³n real

```python
filtrohp=sm.tsa.filters.hpfilter(data["realinv"])
type(filtrohp)
```

```python
realiv_ciclo, relinv_tendencia = sm.tsa.filters.hpfilter(data["realinv"]) #descomponer la tendencia
```

```python
data["tendencia"] = relinv_tendencia
data.head()
```

```python
grafico_3 = plt.figure(figsize=(5,5), dpi=150)
ax = grafico_3.add_axes([0,0,1,1])
ax.plot(data["realinv"], label="InversiÃ³n Real")
ax.plot(data["tendencia"], label="Tendencia de InversiÃ³n")
ax.set_title("InversiÃ³n real en EEUU entre 1959-2009")
ax.set_xlabel("AÃ±o-Trimestre")
ax.set_ylabel("Miles de millones USD")
ax.legend(loc='best') 
ax.grid(True)
plt.show()
```

### Componente Ciclico del Gasto Real

El componente ciclico en el contexto de analisis de series de tiempo se refiere a las fluctuaciones o patrones que se rfepetin en los datos a lo largo del tiempo, pero que no siguen una tencia constante hascia arriba o hacia a bajo.

#### Grafico de componente ciclico

```python
grafico_5 = plt.figure(figsize=(5,5), dpi=150)
ax = grafico_5.add_axes([0,0,1,1])
ax.plot(realiv_ciclo_cons)
ax.set_title("Componente ciclico del gasto real en EEUU entre 1959-2009")
ax.set_xlabel("AÃ±o-Trimestre")
ax.set_ylabel("Miles de millones USD")
ax.grid(True)
plt.show()
```
